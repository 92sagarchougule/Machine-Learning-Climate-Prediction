```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
df = pd.read_csv('Ratnagiri..csv')
df
```

```python
plt.figure(figsize=(9,5))
plt.title(' Average Maximum & Minimum Temperature of Ratnagiri District (1988 to 2018)')
plt.xlabel(' [ Years ]')
plt.ylabel('[ Temperature in Celsius ]')
plt.plot(df.YEAR,df.MAXIMUM_TEMP,label='Maximum Temperature',color='r')
plt.plot(df.YEAR,df.MINIMUM_TEMP,label='Minimum Temperature',color='y')
plt.legend(loc=5)
plt.grid(color='gray', linestyle='-', linewidth=0.4)
plt.savefig('Average Maximum & Minimum Temperature of Ratnagiri District (1988 to 2018).png',transparant=True,b_boxinches='tight', dpi=350)
```

```python
plt.figure(figsize=(9,5))
plt.title('Average Wind Speed of Ratnagiri District (1988 to 2018)')
plt.xlabel('[ Years ]')
plt.ylabel('[ Velocity Meters Per Second (m/s) ]')
plt.plot(df.YEAR,df.AVERAGE_WIND_SPEED,label='Wind Speed',color='b')
plt.legend(loc=1)
plt.grid(color='gray', linestyle='-', linewidth=0.4)
plt.savefig('Average Wind Speed of Ratnagiri District (1988 to 2018).png',transparant=True,b_boxinches='tight', dpi=350)
```

```python
plt.figure(figsize=(9,5))
plt.title('Average Precipitation of Ratnagiri District (1988 to 2018)')
plt.xlabel('[ Years ]')
plt.ylabel('[ In Millimetres (mm) ]')
plt.plot(df.YEAR,df.PRECIPITATION,label='Precipitation',color='g')
plt.legend(loc=2)
plt.grid(color='gray', linestyle='-', linewidth=0.4)
plt.savefig('Average Precipitation of Ratnagiri District (1988 to 2018).png',transparant=True,b_boxinches='tight', dpi=350)
```

# Model

```python
reg = linear_model.LinearRegression()
```

# Equation for calculation of Climate Data

# Max-Temp = M * Year + B

M = reg.coef_

Year = Year of Rainfall

B = reg.intercept_

```python
df1 = pd.read_csv('Year_to_predicted_rainfall.csv')
df1.head()
```

```python
MAXIMUM_TEMP = reg.fit(df[['YEAR']],df.MAXIMUM_TEMP)
Max_Temp = MAXIMUM_TEMP.predict(df1)
```

```python
MAXIMUM_TEMP.coef_       #Coefficient of ---- M value   
```

```python
MAXIMUM_TEMP.intercept_  #  intercept value ---- B value
```

# Mini-Temp= M * Year + B

M = reg.coef_

#Year = Year of Rainfall

#B = reg.intercept_

```python
MINIMUM_TEMP = reg.fit(df[['YEAR']],df.MINIMUM_TEMP)
Min_Temp = MINIMUM_TEMP.predict(df1)
```

```python
MINIMUM_TEMP.coef_  
```

```python
MINIMUM_TEMP.intercept_
```

# Wind-Speed = M * Year + B

```python
AVERAGE_WIND_SPEED = reg.fit(df[['YEAR']],df.AVERAGE_WIND_SPEED)
Avg_Wind_Speed = AVERAGE_WIND_SPEED.predict(df1)
```

```python
#M = reg.coef_

#Year = Year of Rainfall

#B = reg.intercept_
```

```python
AVERAGE_WIND_SPEED.coef_
```

```python
AVERAGE_WIND_SPEED.coef_
```

# Precipitation = M * Year + B

```python
Pre_PRECIPITATION = reg.fit(df[['YEAR']],df.PRECIPITATION)
Precipitation = Pre_PRECIPITATION.predict(df1)
```

```python
#M = reg.coef_

#Year = Year of Rainfall

#B = reg.intercept_
```

```python
Pre_PRECIPITATION.coef_
```

```python
Pre_PRECIPITATION.intercept_
```

```python
df1['Pre_MAXIMUM_TEMP'] = Max_Temp
```

```python
df1['Pre_MINIMUM_TEMP'] = Min_Temp
```

```python
df1['Pre_Avg_Wind_Speed'] = Avg_Wind_Speed
```

```python
df1['Pre_Precipitation'] = Precipitation
```

```python
df1   #Predicted Rainfall Using Single Variable linear Regression
```

```python
plt.figure(figsize=(9,5))
plt.title('Prediction of Maximum & Minimum Temperature of Ratnagiri District (Next 30 year)')
plt.xlabel('Years')
plt.ylabel('Temperature in Celsius')
plt.plot(df.YEAR,df.MAXIMUM_TEMP,label='Maximum Temperature',color='r')
plt.plot(df.YEAR,df.MINIMUM_TEMP,label='Minimum Temperature',color='y') 
plt.plot(df1.YEAR,df1.Pre_MAXIMUM_TEMP,'--',label='Predicted Maximum Temperature',color='r')
plt.plot(df1.YEAR,df1.Pre_MINIMUM_TEMP,'--', label='Predicted Minimum Temperature',color='y')
plt.legend()
plt.grid(color='gray', linestyle='-', linewidth=0.4)
plt.savefig('Prediction of Maximum & Minimum Temperature of Ratnagiri District (Next 30 year).png',transparant=True,b_boxinches='tight', dpi=350)
```

```python
plt.figure(figsize=(9,5))
plt.title('Prediction of Average Wind Speed of Ratnagiri District (Next 30 year) ')
plt.xlabel('[ Years ]')
plt.ylabel(' [ Velocity Meters Per Second (m/s)  ]')
plt.plot(df.YEAR,df.AVERAGE_WIND_SPEED,label='Wind Speed',color='b')
plt.plot(df1.YEAR,df1.Pre_Avg_Wind_Speed,'--g',label='Predicted Wind Speed',color='b')
plt.legend(loc=1)
plt.grid(color='gray', linestyle='-', linewidth=0.4)
plt.savefig('Prediction of Average Wind Speed of Ratnagiri District (Next 30 year).png',transparant=True,b_boxinches='tight', dpi=350)
```

```python
plt.figure(figsize=(9,5))
plt.title('Prediction of Average Precipitation of Ratnagiri District (Next 30 year) ')
plt.xlabel('Years')
plt.ylabel('[ In Millimetres (mm) ]')
plt.plot(df.YEAR,df.PRECIPITATION,label='Precipitation',color='g')
plt.plot(df1.YEAR,df1.Pre_Precipitation,'--r',label='Predicted Precipitation',color='g')
plt.legend(loc=1)
plt.grid(color='gray', linestyle='-', linewidth=0.4)
plt.savefig('Prediction of Average Precipitation of Ratnagiri District (Next 30 year).png',transparant=True,b_boxinches='tight', dpi=350)
```

```python
plt.figure(figsize=(9,5))
plt.title('Prediction of Average Climate of Ratnagiri District (Next 30 year) ')
plt.xlabel('[ Years ]')
plt.ylabel('')
plt.plot(df1.YEAR,df1.Pre_Precipitation,'--',label='',color='g')
plt.plot(df1.YEAR,df1.Pre_Avg_Wind_Speed,'--',label='',color='b')
plt.plot(df1.YEAR,df1.Pre_MAXIMUM_TEMP,'--',label='',color='r')
plt.plot(df1.YEAR,df1.Pre_MINIMUM_TEMP,'--', label='',color='y')
plt.plot(df.YEAR,df.PRECIPITATION,label='Precipitation',color='g')
plt.plot(df.YEAR,df.MAXIMUM_TEMP,label='Maximum Temperature',color='r')
plt.plot(df.YEAR,df.MINIMUM_TEMP,label='Minimum Temperature',color='y') 
plt.plot(df.YEAR,df.AVERAGE_WIND_SPEED,label='Wind Speed',color='b')
plt.legend(loc=6)
plt.grid(color='gray', linestyle='-', linewidth=0.4)
plt.savefig('Prediction of Average Climate of Ratnagiri District (Next 30 year).png',transparant=True,b_boxinches='tight', dpi=350)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
df = pd.read_excel('Seasonal_Precipitation.xlsx','Sindhudurg')
df1 = pd.read_excel('Seasonal_Precipitation.xlsx','Ratnagiri')
df2 = pd.read_excel('Seasonal_Precipitation.xlsx','Raigad')
```

```python
plt.figure(figsize=(9,5))
plt.title('Seasonal Precipitation of Konkan Region, Maharashtra(June to September)')
plt.xlabel('[ Years ]')
plt.ylabel('[ In Millimetres (mm) ]')
plt.plot(df.YEAR,df.PRECIPITATION,label='Sindhudurg District', color='g')
plt.plot(df1.YEAR,df1.PRECIPITATION,label='Ratnagiri District',color='m')
plt.plot(df2.YEAR,df2.PRECIPITATION,label='Raigad District', color='b')
plt.legend(loc=2)
plt.grid(color='gray', linestyle='-', linewidth=0.4)
plt.savefig('Seasonal Precipitation (1988 to 2018).png',transparant=True,b_boxinches='tight', dpi=350)
```

# Model

```python
reg = linear_model.LinearRegression()
```

```python
df_ = pd.read_csv('Year_to_predicted_rainfall.csv')
df1_ = pd.read_csv('Year_to_predicted_rainfall.csv')
df2_ = pd.read_csv('Year_to_predicted_rainfall.csv')
```

# (Sindhudurg)Precipitation = M * Year + B

```python
Pre_PRECIPITATION = reg.fit(df[['YEAR']],df.PRECIPITATION)
Precipitation = Pre_PRECIPITATION.predict(df_)
```
